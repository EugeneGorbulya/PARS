# 1. Краткое описание

**SmartRentRanker** — это система, которая превращает поиск аренды жилья в персональный рекомендательный сервис.  
Она собирает объявления с **Циана**, анализирует фотографии и характеристики квартир,  
а пользователь помогает системе обучиться своему вкусу.

Пользователь через Telegram-бота оценивает квартиры по трём факторам (по шкале от 1 до 5):

- **Красота (*beauty*)** — насколько визуально нравится квартира.  
- **Цена-качество (*price_quality*)** — насколько квартира кажется выгодным вариантом.  
- **Расположение (*distance_preference*)** — насколько удобно добираться до нужной точки.

После того как пользователь оценит примерно **80–120 квартир**, система обучает модель,  
которая начинает предсказывать его предпочтения и формировать персональный рейтинг объявлений.  
Ранжирование строится с учётом фильтров, цены и времени в пути.

*Результат — поиск квартир «по вкусу», а не по формальным параметрам.*


# 2. End-to-End сценарий

1. **Пользователь запускает Telegram-бота.**  
   Бот объясняет идею: «Я помогу найти квартиры, которые тебе действительно нравятся — по красоте, цене и расположению.»

2. **Создание профиля (alias).**  
   Пользователь задаёт фильтры поиска: город, диапазон цен, количество комнат, а также указывает точку, до которой важно время в пути (например, место работы).  
   Эти фильтры сохраняются и используются для формирования запросов к API Циана.

3. **Сбор данных.**  
   Система обращается к публичному API Циана и получает список актуальных объявлений, подходящих под фильтры профиля.  
   Для каждого адреса дополнительно вычисляется время в пути через API Яндекс.Карт.

4. **Показ квартир пользователю.**  
   Бот отправляет **карточки квартир** с краткой сводкой:
   - цена, площадь, адрес, ближайшее метро;
   - время в пути до указанной точки (POI);
   - **10 превью-фотографий** (обложки объявления);
   - **ссылку на полное объявление на Циан**, где пользователь может просмотреть все фотографии.

   Пользователь оценивает квартиру **по фотографиям с Циана**,  
   выставляя баллы по трём факторам (1–5):
   - красота (**beauty**),
   - цена–качество (**price_quality**),
   - расположение (**distance_preference**).

   Если объявление невозможно оценить (например, только планировка или повтор),  
   его можно **пропустить**, и оно не будет учитываться в обучении.


5. **Накопление данных и обучение.**  
   Когда пользователь поставит достаточно оценок (около 80–120 квартир), система обучает модель предпочтений.  
   Модель использует визуальные эмбеддинги (CLIP) и табличные признаки (цена, площадь, этаж, расстояние и т.д.),  
   чтобы предсказывать вероятную оценку пользователя по каждому фактору.

6. **Рассчёт итогового рейтинга.**  
   Для каждой квартиры система вычисляет сводный показатель привлекательности (*score*),  
   комбинируя предсказанные оценки по факторам с весами, заданными пользователем.  
   Квартиры сортируются по этому score — от наиболее подходящих к наименее.

7. **Выдача рекомендаций.**  
   Пользователь получает список лучших квартир, может изменить веса критериев  
   (например, сделать расположение важнее красоты) или переключиться на другой профиль.

8. **Инкрементальное обновление.**  
   Система периодически обращается к Циану, добавляет новые объявления и показывает пользователю только те квартиры, которых он ещё не видел.

---

📈 *В результате SmartRentRanker превращает обычный поиск по фильтрам  
в персональный рекомендательный процесс, который подстраивается под вкус конкретного человека.*


# 3. Архитектура (высокоуровнево)

Система SmartRentRanker состоит из нескольких взаимосвязанных компонентов,  
которые обеспечивают сбор данных, обработку, обучение модели и взаимодействие с пользователем.

---

## Основные компоненты

- **Telegram Bot** — интерфейс взаимодействия с пользователем.  
  Создаёт и управляет профилями, показывает квартиры, принимает оценки  
  и выдаёт персональные рекомендации.  
  После первого этапа обучения бот переходит в режим **парных сравнений (pairwise)**,  
  где пользователь выбирает, какая из двух квартир ему больше нравится **по конкретному фактору**:  
  - по красоте,  
  - по соотношению цена–качество,  
  - или по удобству расположения.  
  Это помогает уточнить границы предпочтений по каждому фактору отдельно.

- **Fetcher** — сервис сбора данных из публичного API Циана.  
  Выполняет постраничные запросы, фильтрует результаты, удаляет дубликаты  
  и сохраняет новые объявления в базу данных.

- **Enricher** — сервис обогащения данных.  
  Добавляет время в пути через API Яндекс.Карт, определяет типы помещений по фото,  
  извлекает визуальные эмбеддинги с помощью модели CLIP.

- **ML Service** — модуль обучения предпочтений.  
  Обучает многозадачную модель с несколькими «головами»  
  (по факторам: *beauty*, *price_quality*, *distance_preference*).  
  Поддерживает два режима обучения:
  1. **Single-label** — обучение на индивидуальных оценках 1–5 по каждому фактору.  
  2. **Pairwise** — дообучение на парных сравнениях по выбранному фактору  
     (“какая квартира лучше по цене?”, “где расположение удобнее?”, “какая красивее?”).  
  ML-сервис хранит чекпойнты моделей в S3 и обновляет метрики стабильности (Kendall τ, MAE).

- **Ranker** — пересчитывает итоговый рейтинг квартир.  
  Комбинирует предсказания модели с пользовательскими весами факторов  
  и формирует итоговый *score*.

- **Хранилища данных:**  
  - **PostgreSQL** — основное хранилище (квартиры, пользователи, оценки, фильтры).  
  - **MinIO / S3** — хранение изображений и чекпойнтов моделей.  
  - **Redis** — кэш и очереди (например, travel-time и новые объявления).

---

## Поток взаимодействия


Пользователь → Telegram Bot → Fetcher → Enricher → ML Service → Ranker → Telegram Bot


---

📦 *Такая архитектура позволяет независимо масштабировать каждый компонент  
и добавлять новые источники данных или факторы без изменения всей системы.*


# 4. Данные и профили

В основе SmartRentRanker лежит понятие **профиля (alias)** —  
это персональный набор фильтров, предпочтений и оценок пользователя,  
на основании которых система обучает модель вкуса и строит ранжирование квартир.

---

## Структура данных

- **Профиль (alias)**  
  Содержит:
  - фильтры Циана (город, диапазон цен, комнаты, этаж, тип жилья);
  - координаты точки интереса (например, работы);
  - веса факторов (w_b, w_pq, w_d), определяющие важность критериев:
    - красота (beauty),
    - цена–качество (price_quality),
    - расположение (distance_preference);
  - состояние обучения (кол-во меток, стадия: single/pairwise, стабильность модели);
  - id последнего обученного чекпойнта модели.

---

- **Квартира (flat)**  
  Представляет объявление с Циана и содержит:
  - id Циана, цена, площадь, этаж, адрес, координаты;
  - район и ближайшие станции метро;
  - фотографии и ссылки на оригинал;
  - travel_time (минуты до точки интереса по API Яндекс.Карт);
  - CLIP-эмбеддинги фотографий;
  - служебные поля: дата добавления, актуальность, источник.

---

- **Оценка (rating)**  
  Одиночные оценки пользователя (по шкале 1–5):
user_id, profile_alias, flat_id, beauty, price_quality, distance, skipped

Используются для обучения базовых голов модели (single-label режим).

---

- **Сравнение (pairwise_rating)**  
Используется при парном режиме обучения:

user_id, profile_alias, flat_a, flat_b, factor, preferred_flat

Каждая запись означает выбор квартиры A или B по указанному фактору  
(например, "A красивее", "B ближе к работе").

---

- **Кэш (cache)**  
- просмотренные `flat_id` — чтобы не показывать повторно;  
- travel-time — для снижения числа обращений к API Яндекс.Карт;  
- глобальный кэш квартир и эмбеддингов — общий для всех пользователей.

---

## Жизненный цикл профиля

1. Пользователь создаёт профиль и задаёт фильтры.  
2. Fetcher собирает квартиры, удовлетворяющие фильтрам.  
3. Пользователь оценивает часть квартир → появляются записи в `rating`.  
4. После достижения порога меток модель обучается (single-label).  
5. При достижении стабильности — переход к pairwise-режиму и уточнение по факторам.  
6. Модель сохраняется как чекпойнт (в S3), профиль помечается как “обученный”.  
7. Пользователь может переключаться между профилями, менять веса или использовать чужие (популярные) профили.

---

🧠 *Профиль объединяет фильтры, оценки и модель предпочтений пользователя —  
он представляет собой персональную “точку зрения” на рынок аренды.*


# 5. Pipeline обучения (c Exploration-mode внутри этапов)

SmartRentRanker обучает вкус пользователя в два этапа:  
сначала по одиночным оценкам (*Single-label*), затем уточняет результат с помощью парных сравнений (*Pairwise*).  
Каждый фактор — **красота**, **цена–качество**, **расположение** — обучается как отдельная «голова» модели.  
Во время обоих этапов используется **Exploration-mode**: контролируемая подмешка “неуверенных/новых” объявлений, чтобы ускорять обучение и избегать зацикливания.

---

## Этап A — Single-label (оценки 1–5)

**Цель:** быстро понять базовые предпочтения по каждому фактору.

- **Данные:** таблица `rating` (оценки 1–5).
- **Модель:** общий backbone (CLIP + табличные признаки) и три головы: `beauty`, `price_quality`, `distance_preference`.
- **Лосс:** среднее по головам (учитываются только непустые факторы).
- **Балансировка:** собираем низкие/средние/высокие оценки; избегаем перекоса “все плохие” или “все средние”.
- **Exploration-mode (на этапе A):**
  - **ε-greedy:** 70–90% показов по текущему score (exploit), 10–30% — *explore*.  
    Рекомендуем: ε=0.2 на старте, снижаем до 0.1 после стабилизации (см. ниже).
  - **Что берём в explore:**
    1) «неуверенные» (близкие по score кандидаты / высокая дисперсия предсказаний),  
    2) «свежие» (новые объявления),  
    3) «непокрытые зоны» (районы/диапазоны цен или типы планировок, где мало меток).
  - **Диверсификация:** к top-кандидатам применяем MMR (снижаем похожесть соседей по косинусной близости эмбеддингов).
  - **UX-ограничения:** не более 1–2 exploration-показов подряд.
- **Порог перехода на B:**
  - 30–50 оценок на фактор (≈ 80–120 квартир суммарно),
  - сбалансированное распределение меток,
  - стабильность топ-20 по каждому фактору (Kendall τ ≥ 0.8) на двух подряд обучениях.
- При выполнении условий сохраняем чекпойнт и переводим профиль в **pairwise**.

---

## Этап B — Pairwise (сравнения «A лучше B»)

**Цель:** уточнить границы предпочтений по каждому фактору.

- **Данные:** таблица `pairwise_rating` с выбором по конкретному фактору  
  (например, «A красивее», «B выгоднее», «A удобнее по расположению»).
- **Выбор пар:**
  - кандидаты из «пограничной» зоны (позиции 5–30 текущего ранжирования),
  - разнообразие по районам/типам, чтобы избежать однообразия.
- **Лосс:** pairwise-hinge или logistic (вероятность, что выбранная квартира должна иметь больший предсказанный рейтинг по фактору).
- **Дообучение:** короткие сессии (few-shot) с ранней остановкой и регуляризацией; обновляем только головы.
- **Exploration-mode (на этапе B):**
  - ε-greedy для **выбора пар**: 80–90% — «наиболее спорные» пары (минимальная маржа), 10–20% — пары из зон, где у фактора мало меток.  
  - Фокус-фактор: если система хочет «прокачать» конкретный фактор, повышаем долю пар именно по нему.
  - MMR к списку кандидатов для пар, чтобы пары не были однотипными.
  - UX-ограничения: не более 2 “сложных” пар подряд.
- **Адаптация ε:** снижаем ε при росте стабильности; повышаем ε, если модель “застыла” или у фактора низкое покрытие.

---

## Контроль переобучения и стабильности

Запуск каждые несколько итераций обучения:

- **Hold-out по времени (train/val split):** если train↓ и val↑ — стоп/усилить регуляризацию.
- **Стабильность топ-20:** не должна падать ниже 0.7; при снижении — приостанавливаем обучение, увеличиваем exploration и диверсификацию.
- **Калибровка шкал 1–5:** сравнение распределений предсказаний и реальных оценок (перефит весов при дрейфе).
- **UX-сигналы:** если растёт «frustration proxy» (много пропусков/скрытий подряд), временно уменьшаем ε.

---

## Итоговый расчёт рейтинга

После обучения система предсказывает для каждой квартиры:
- `beautŷ` — визуальная привлекательность,
- `price_qualitŷ` — ощущение выгодности,
- `distancê` — удобство расположения.

Итоговый **score**:

$$
\text{score}_i =
w_b \cdot \hat{beauty}_i +
w_{pq} \cdot \hat{price\_quality}_i +
w_d \cdot \hat{distance}_i
- \text{penalties}_i
$$

где $(w_b, w_{pq}, w_d)$ — веса профиля, а $\text{penalties}_i$ — штрафы (например, устаревшее объявление).

---

## Диаграмма потока обучения (c Exploration)

$$
\begin{array}{c}
\boxed{\textbf{Оценки 1–5 (beauty, price, distance)}} \\[8pt]
\downarrow \\[8pt]
\boxed{\textbf{Этап A: Single-label (3 головы)}} \\[4pt]
\textit{Exploit: ранжированный список} \quad|\quad
\textit{Explore: ε, свежие, неуверенные, MMR} \\[4pt]
\text{Критерии перехода: покрытие + стабильность топ-20 } (\tau \ge 0.8) \\[10pt]
\downarrow \\[8pt]
\boxed{\textbf{Этап B: Pairwise A/B по выбранному фактору}} \\[4pt]
\textit{Exploit: спорные пары} \quad|\quad
\textit{Explore: пары из непокрытых зон, MMR} \\[10pt]
\downarrow \\[8pt]
\boxed{\textbf{Дообучение голов (fine-tuning), контроль валидации и стабильности}} \\[10pt]
\downarrow \\[8pt]
\boxed{\textbf{Финальная модель профиля} \Rightarrow \textbf{ Score и ранжирование}} \\
\end{array}
$$

---

🧠 *Встроенный Exploration-mode в обоих этапах даёт системе «зрячее» обучение:  
она быстро набирает информативные метки, избегает зацикливания и стабильнее удерживает качество рекомендаций.*


# 6. Итоговый score и ранжирование (с учётом Exploration)

После того как модель обучена на данных пользователя,  
каждая новая квартира получает прогноз по трём факторам:

- **beautŷ** — визуальная привлекательность;  
- **price_qualitŷ** — ощущение выгодности;  
- **distancê** — удобство расположения (по времени в пути).  

Эти предсказания объединяются в единый интегральный **score**,  
отражающий субъективное восприятие привлекательности квартиры.

---

## Расчёт итогового score

$$
\text{score}_i =
w_b \cdot \hat{beauty}_i +
w_{pq} \cdot \hat{price\_quality}_i +
w_d \cdot \hat{distance}_i
- \text{penalties}_i
$$

где:

- $( w_b, w_{pq}, w_d )$ — веса факторов, заданные пользователем (важность красоты, выгоды и расположения);
- $\text{penalties}_i$ — штрафы за нежелательные признаки (устаревшее объявление, этаж, плохое состояние и т.п.);
- все предсказания нормализованы в диапазоне [0, 1].

---

## Алгоритм ранжирования

1. Для каждой квартиры вычисляется `score_i`.  
2. Квартиры сортируются по убыванию `score_i`.  
3. Применяются пост-корректировки:  
   - исключаются дубликаты и неактуальные объявления,  
   - штрафуются объявления с низкой уверенностью модели.  
4. Формируется финальный список для пользователя.

---

## Exploration при ранжировании

Чтобы поддерживать обучение и избегать стагнации вкуса,  
в итоговый список добавляется **Exploration-доля** (10–20 %) объявлений:

- **Неуверенные** (высокая дисперсия по головам модели);  
- **Свежие** (новые, ещё не оцененные квартиры);  
- **Редкие зоны** (районы или типы квартир, где модель имеет мало данных).  

Алгоритм подбора:
- базовый ε-greedy (ε ≈ 0.1–0.2);
- диверсификация через MMR — минимизация схожести эмбеддингов между соседями.

Итог: пользователь видит в основном надёжно релевантные варианты,  
но иногда — “новые” квартиры, которые помогают модели уточнять вкус.

---

## Пример визуализации

$$
\begin{array}{l|c|c|c|c}
\text{Квартира} & \hat{beauty} & \hat{price\_quality} & \hat{distance} & \text{Score} \\
\hline
\text{A (ЦАО, 95 000 ₽)} & 0.92 & 0.70 & 0.85 & 0.84 \\
\text{B (СЗАО, 60 000 ₽)} & 0.78 & 0.90 & 0.72 & 0.80 \\
\text{C (ЮАО, 45 000 ₽)} & 0.60 & 0.95 & 0.55 & 0.70 \\
\end{array}
$$

---

🧩 *Exploration-вставка на этапе ранжирования позволяет системе не “застывать”  
и находить новые квартиры, способные изменить или уточнить восприятие вкуса пользователя.*


# 7. UX Telegram-бота (финальная версия с Exploration)

Интерфейс Telegram-бота построен так, чтобы пользователь с минимальными действиями
создал профиль, обучил модель вкуса и получил список квартир, максимально близких его предпочтениям.

---

## 7.1 Онбординг

- Приветствие:  
  «Я помогу найти квартиры, которые тебе действительно нравятся — по красоте, цене и расположению.»  
- Кнопки: **Создать профиль** · **Как это работает**.

---

## 7.2 Создание профиля (alias)

1. Пользователь задаёт **базовые фильтры поиска**:  
   город, диапазон цен, количество комнат, этаж и т.д.  
2. Указывает **точку интереса (POI)** и максимальное допустимое время в пути.  
3. Настраивает **веса факторов**:  
   - Красота (**beauty**)  
   - Цена–качество (**price_quality**)  
   - Расположение (**distance_preference**)  
4. Придумывает имя профиля (например, `work_45min`).  
5. (опционально) Помечает профиль как **публичный**, чтобы другие пользователи могли им пользоваться.

После сохранения профиль становится активным:  
все будущие оценки, сравнения и рекомендации будут учитывать выбранные фильтры и веса.

---

### 🔁 Публичные и копируемые профили

Любой профиль можно:
- **Опубликовать** — сделать доступным другим пользователям;  
- **Форкнуть** — создать свою копию с теми же весами и моделью вкуса,  
  но с личными фильтрами и POI;  
- **Скрыть** — убрать из публичного списка.

> При копировании чужого профиля сохраняются веса и обученные головы  
> (beauty, price_quality, distance),  
> но не передаются личные оценки или история просмотров.  

Для нового пользователя публичный профиль может служить готовым “вкусом”,  
который можно использовать **как есть** или **дообучить под себя**.


---

## 7.3 Просмотр и оценка квартир

Команда `/next` запускает показ квартир, полученных через API Циана.

Карточка содержит:
- фотографии (1–3 шт.),
- цену, площадь, адрес, ближайшее метро,
- время в пути до POI.

Под карточкой — кнопки-оценки **1 … 5** по факторам:

- Красота  
- Цена-качество  
- Расположение  

а также действия **Пропустить** / **Открыть на Циан** / **Следующая квартира**.  

После оценки бот иногда показывает “новую” или “неуверенную” квартиру в рамках Exploration —  
для уточнения вкуса и улучшения модели.

---

## 7.4 Прогресс обучения

Периодически бот показывает:

- покрытие шкал (есть ли и “плохие”, и “хорошие” оценки);  
- стабильность топ-20 по каждому фактору (Kendall τ);  
- подсказки:  
  - «Осталось ≈ 10 оценок до точных рекомендаций.»  
  - «Добавил несколько новых квартир, чтобы лучше понять твой вкус.»

---

## 7.5 Переход к парным сравнениям

Когда достигнуто достаточное покрытие и стабильность,  
бот предлагает перейти в **режим сравнения квартир**:

> «Хочешь уточнить мой прогноз? Давай сравним пары квартир.»

Пользователь выбирает фактор:  
**Красота**, **Цена-качество** или **Расположение**.  

Бот показывает две карточки (A и B) и вопрос:  
> «Какая квартира **лучше по выбранному критерию**?»  

Кнопки: **A лучше**, **B лучше**, **Пропустить**.  

Эти сравнения используются для дообучения модели.

---

## 7.6 Топ-рекомендации

Команда `/top` показывает лучшие квартиры,  
отсортированные по персональному score профиля.

Формат карточки:
- фото, цена, адрес, время в пути;  
- краткий рейтинг по факторам;  
- общий score;  
- кнопки: **Открыть на Циан**, **Сохранить**, **Скрыть**.

Также возможен лёгкий **exploration в выдаче** —  
примерно 10–15 % списка составляют новые или неуверенные объявления,  
чтобы уточнять вкус даже во время просмотра рекомендаций.

---

## 7.7 Управление профилями

Команда `/profiles`:
- показывает список профилей (активный отмечен);
- позволяет: **Активировать**, **Редактировать**, **Клонировать**, **Удалить**, **Переименовать**;
- менять веса факторов и параметры POI прямо в меню.

---

## 7.8 Популярные профили

Раздел **Популярные профили** даёт доступ к заранее обученным “вкусам”:
- карточка профиля: город, диапазон цен, количество квартир в обучении, метрика стабильности;
- действия: **Использовать** или **Форкнуть** (создать свою копию и дообучить).

---

## 7.9 Умные подсказки и Exploration

- Если оценки однородные (все “плохие” или все “средние”) —  
  предложение **включить exploration** или расширить фильтры (цена/район/время).  
- Если очередь пуста: «Ищу новые объявления… нашёл N новых.»  
- Если модель нестабильна: «Лучше ещё 10 оценок по расположению.»

---

## 7.10 Команды кратко

- `/start` — онбординг  
- `/new_profile` — создать профиль  
- `/profiles` — список и управление профилями  
- `/next` — следующая квартира  
- `/compare` — парные сравнения  
- `/top` — показать лучшие квартиры  
- `/weights` — изменить веса факторов  
- `/poi` — изменить точку интереса  
- `/settings` — прочие настройки  

---

🎛️ *UX построен так, чтобы обучение происходило естественно:  
бот подмешивает немного exploration, но основной поток — релевантные и понятные квартиры.*


# 8. Архитектура проекта

Система SmartRentRanker разделена на независимые компоненты,  
чтобы обеспечить масштабируемость, кэширование и простоту обновления моделей.

---

## 8.1 Общая схема

$$
\begin{array}{c}
\boxed{\textbf{Telegram Bot}}\\
\text{Интерфейс пользователя, команды /next /top /compare}\\[8pt]
\downarrow\\[8pt]
\boxed{\textbf{Backend API (FastAPI)}}\\
\text{REST/gRPC слой, фильтры, профили, обучение}\\[8pt]
\downarrow\\[8pt]
\boxed{\textbf{PostgreSQL Database}}\\
\text{users, profiles, ratings, pairwise, flats, cache}\\[8pt]
\downarrow\\[8pt]
\boxed{\textbf{ML Engine (Python)}}\\
\text{CLIP + табличные признаки, 3 головы, fine-tuning, инференс}\\[8pt]
\downarrow\\[8pt]
\boxed{\textbf{CIAN / Яндекс API}}\\
\text{Источники квартир и расчёт времени в пути}\\
\end{array}
$$


---

## 8.2 Поток данных

1. **Пользователь → Telegram Bot**  
   вводит фильтры, оценивает квартиры, создаёт профиль.  

2. **Bot → Backend API**  
   запросы `/next`, `/rate`, `/compare`, `/top`.  
   API выбирает квартиры из БД или кэша, применяя фильтры профиля.

3. **Backend → ML-модуль**  
   при обучении — отправляет новые метки (`rating`, `pairwise`)  
   при инференсе — запрашивает score для новых квартир.

4. **ML-модуль → PostgreSQL**  
   сохраняет обновлённые веса профиля, стабильность, метрики.

5. **CIAN API → Backend**  
   периодически подтягивает новые квартиры, обновляет кэш,  
   фильтрует по городу и дате публикации.

---

## 8.3 Асинхронность и кэширование

- Все сетевые операции (CIAN, Яндекс API, инференс) — асинхронные.  
- Квартиры, однажды просмотренные пользователем, помечаются в БД и больше не предлагаются.  
- Redis (опционально) используется как слой кэша для свежих объявлений и коротких TTL (1–2 ч).  
- Модуль обучения выполняется в фоне (Celery или отдельный воркер).

---

## 8.4 Разделение логики

| Компонент | Ответственность |
|------------|----------------|
| **Telegram Bot** | UX, диалог, сбор оценок |
| **Backend API** | маршрутизация, фильтры, управление профилями |
| **PostgreSQL** | хранилище квартир, оценок и прогресса обучения |
| **ML Engine** | обучение и предсказание по изображениям и табличным данным |
| **CIAN / Яндекс API** | внешние источники квартир и расчёта времени в пути |

---

📦 *Такое разбиение позволяет независимо обновлять модели,  
кэшировать внешние запросы и легко масштабировать бота под несколько городов или пользователей.*


# 9. Технологический стек

Система SmartRentRanker построена на современном, лёгком и воспроизводимом стеке.  
Основная цель — простота разработки, масштабируемость и лёгкая интеграция ML-моделей.

---

## 9.1 Языки и фреймворки

| Компонент | Технологии | Назначение |
|------------|-------------|------------|
| **Backend API** | 🐍 **Python**, **FastAPI**, **Pydantic** | асинхронный API-слой, управление профилями и фильтрами |
| **Бот** | **aiogram 3**, **Telebot (альтернатива)** | диалог с пользователем, обработка команд и inline-кнопок |
| **ML-модуль** | **PyTorch**, **CLIP**, **Darts/LightGBM** | обучение и инференс визуальных и табличных моделей |
| **База данных** | **PostgreSQL 15** | хранение квартир, фильтров, профилей, оценок, результатов инференса |
| **Кэш и очередь** | **Redis**, **Celery** | кэширование свежих объявлений, асинхронное обучение и обновления |
| **Парсер Циана** | **HTTPX / Aiohttp** | асинхронные запросы к Public API Циана, обновление кэша |
| **Гео-API** | **Yandex Maps API / Distance Matrix** | расчёт времени в пути до заданной точки (POI) |
| **DevOps** | **Docker + Docker Compose**, **Uvicorn**, **NGINX** | изоляция сервисов и деплой |
| **Хранение моделей** | **S3 (minio / AWS)** | сохранение чекпойнтов профилей и эмбеддингов |

---

## 9.2 Машинное обучение

**Модель визуального восприятия**
- Базируется на `CLIP (ViT-B/32)` с fine-tuning на пользовательских оценках.
- Добавляется “табличная голова” для числовых признаков квартиры (цена, площадь, этаж).
- Три независимых головы: `beauty`, `price_quality`, `distance_preference`.
- Используется multi-task обучение с взвешенным лоссом.

**Pairwise-обучение**
- Дообучение с hinge/logistic-loss по парам “A лучше B”.
- Контроль переобучения по стабильности топ-20 (Kendall τ).

**Exploration-режим**
- ε-greedy (ε ≈ 0.1–0.2) + MMR для диверсификации выдачи.  
- Асинхронный сбор новых меток и дообучение фоном.

---

## 9.3 Архитектура хранения данных

- **PostgreSQL**
  - `users`, `profiles`, `ratings`, `pairwise`, `flats`, `cache`
  - JSONB-поля для гибких фильтров и параметров профиля  
- **Redis**
  - TTL-кэш свежих квартир (1–2 ч)
  - счётчики и флаги exploration  
- **S3**
  - чекпойнты моделей, эмбеддинги CLIP-векторов  
- **CIAN API + Yandex Maps API**
  - обновление данных раз в N минут  
  - вычисление travel-time feature  

---

## 9.4 Разработка и развертывание

- локальная разработка → `docker-compose up` (API + DB + Redis)  
- CI/CD → GitLab CI (build, test, deploy)  
- деплой: **Uvicorn + Nginx**
- планировщик обновлений квартир и переобучений: **Celery beat / Cron**

---

🧩 *Такой стек обеспечивает баланс между скоростью прототипирования,  
асинхронной обработкой данных и возможностью последующего масштабирования под реальный продакшен.*

где каждый пользователь может не только искать, но и делиться своим восприятием красоты и уюта.*


---

### users

* **Кто:** учетные записи людей (привязка к Telegram).
* **Зачем:** идентифицировать автора профилей/оценок, хранить базовые метаданные.

### profiles

* **Кто:** персональные «вкусы» (alias) внутри пользователя.
* **Что хранит:** фильтры Циана, веса факторов, стадия обучения (single/pairwise), ε для exploration, публичность/slug, от кого форкнут, последний чекпойнт.
* **Зачем:** по профилю считаем score и строим выдачу; один пользователь может иметь несколько профилей (работа/дача и т.п.).

### pois

* **Кто:** точки интереса пользователя (работа, зал и т.д.).
* **Зачем:** базовые координаты для расчёта времени в пути; могут использоваться в нескольких профилях.

### profile_pois

* **Кто:** связь профиль ↔ POI.
* **Что хранит:** лимит минут до POI, режим маршрута, приоритет.
* **Зачем:** один профиль может учитывать несколько POI (например, работа и садик) с разными порогами.

### flats

* **Кто:** квартиры/объявления из Циана (нормализованный оффер).
* **Что хранит:** cian_id, цена, метры, адрес/координаты, этаж, материал, метро, дата публикации, активность.
* **Зачем:** основной каталог объектов для ранжирования и показа.

### flat_photos

* **Кто:** фото конкретной квартиры.
* **Что хранит:** порядок, URL, предполагаемый тип помещения (room_type).
* **Зачем:** отображение в боте, извлечение визуальных признаков, фильтр «неоценимые» (план/дубликат).

### photo_embeddings

* **Кто:** метаданные по эмбеддингам фото (вектор лежит в S3).
* **Зачем:** быстро находить ссылку на вектор + версию модели/размерность; не раздувать БД бинарниками.

### flat_poi_travel

* **Кто:** кэш времени в пути квартира ↔ POI.
* **Что хранит:** минуты, режим, TTL.
* **Зачем:** не вызывать внешнее API для каждой выдачи; ускоряет score по фактору «distance».

### ratings

* **Кто:** одиночные оценки пользователя по квартире в рамках профиля.
* **Что хранит:** beauty, price_quality, distance (1–5), skipped.
* **Зачем:** основной таргет для **single-label** обучения; по (user, profile, flat) одна запись.

### pairwise_ratings

* **Кто:** парные сравнения A vs B по выбранному фактору.
* **Зачем:** таргеты для **pairwise** дообучения (тонкая настройка границ предпочтений).

### profile_delivery_queue

* **Кто:** очередь кандидатов на показ профилю.
* **Что хранит:** состояние (queued/shown/rated/skipped).
* **Зачем:** управлять порядком выдачи, повторно не подсовывать показанное, смешивать exploration.

### seen_flats

* **Кто:** факт, что пользователь уже видел квартиру в профиле.
* **Зачем:** не показывать повторно; база для «только новые».

### saved_flats

* **Кто:** «сохранённые» пользователем квартиры.
* **Зачем:** быстрый доступ к понравившимся, повышающий сигнал релевантности.

### hidden_flats

* **Кто:** скрытые пользователем квартиры.
* **Зачем:** исключить из будущих выдач; сильный негативный сигнал.

### model_snapshots

* **Кто:** чекпойнты обученной модели профиля.
* **Что хранит:** тип backbone/голов, URI в S3, метрики (в т.ч. Kendall τ, MAE), время создания.
* **Зачем:** версионирование и откат; шаринг/форк используют конкретный снапшот.

### profile_metrics

* **Кто:** агрегированные метрики по профилю.
* **Что хранит:** кол-во single/pairwise меток, текущую стабильность, время последнего обучения.
* **Зачем:** показывать прогресс, решать о переходе на pairwise/изменении ε.

### profile_flat_score

* **Кто:** материализованный кэш score для (профиль, квартира).
* **Что хранит:** общий score + предсказания по факторам, время обновления.
* **Зачем:** быстрый `/top` без онлайнового инференса по всему пулу; удобно инкрементально обновлять.
